## Imprting Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset



```python
dataset = pd.read_csv('Student_Marks.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test *set*

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

## Training the Polynomial Regression model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 8)
X_poly = poly_reg.fit_transform(x_train)
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly,y_train)
```

## Visualizing Multiple Linear Regrerssion -Training Data


```python
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, lin_reg.predict(x_train), color = "blue" )
plt.title("Salary vs Experience (Training set)")
plt.xlabel("years of Experience")
plt.ylabel("Salary")
```

## Visualizing Multiple Linear Regrerssion -Testing Data


```python
plt.scatter(x_test, y_test, color = "red")
plt.plot(x_train, lin_reg.predict(x_train), color = "blue" )
plt.title("Salary vs Experience (Training set)")
plt.xlabel("years of Experience")
plt.ylabel("Salary")
```

```python
# Create the first plot for Polynomial Linear Regression (Training set)
plt.figure(figsize=(12, 6))

# Scatter plot for Training set
plt.subplot(1, 2, 1)
plt.scatter(x_train, y_train, color="red")

# Plot Polynomial Linear Regression (curved line) for Training set
X_train_poly = poly_reg.transform(x_train)  # Apply polynomial transformation to x_train
plt.plot(x_train, lin_reg2.predict(X_train_poly), color="blue")
plt.title("Polynomial Linear Regression (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

# Create the second plot for Polynomial Linear Regression (Test set)
plt.subplot(1, 2, 2)
plt.scatter(x_test, y_test, color="red")

# Plot Polynomial Linear Regression (curved line) for Test set
X_test_poly = poly_reg.transform(x_test)  # Apply polynomial transformation to x_test
plt.plot(x_test, lin_reg2.predict(X_test_poly), color="blue")
plt.title("Polynomial Linear Regression (Test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

# Display the plots
plt.tight_layout()  # Adjust subplots to prevent overlap
plt.show()
```

## Accuracy of Multiple Linear Regression

```python
from sklearn.metrics import mean_squared_error, r2_score
y_pred_1 = lin_reg.predict(x_test)

# R-squared (R²)
r2 = r2_score(y_test, y_pred_1)
print(f"R-squared (R²): {r2:.4f}")

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_1)
print(f"Mean Squared Error (MSE): {mse:.4f}")

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

## Accuracy of polynomial regression

```python
y_pred_2 = lin_reg2.predict(poly_reg.transform(x_test))

# R-squared (R²)
r2 = r2_score(y_test, y_pred_2)
print(f"R-squared (R²): {r2:.4f}")

# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_2)
print(f"Mean Squared Error (MSE): {mse:.4f}")

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```
